```python
import requests
import re
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import csv
```

```python
url = "https://www.nature.com/articles/s41598-023-28880-x"
page = urlopen(url)
html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
type(soup)
```

```python
print(soup)
```

DISPLAYING THE TITLE OF THE ARTICLE 

```python
title=soup.find_all("h1")
pattern = "<h1.*?>.*?</h1.*?>"
match_results = re.search(pattern,html, re.IGNORECASE)
title = match_results.group()
title = re.sub("<.*?>", "",title)
print(title)
```

```python
title=soup.find_all("h1")
print(title[0].text)
```

DISPLAYING THE ABSTRACT OF THE ARTICLE

```python
abs=soup.find_all("div",class_="main-content")
pattern = "<div.*?>.*?</div.*?>"
match_results = re.search(pattern, html, re.IGNORECASE)
abs = match_results.group()
abs= re.sub("<.*?>", "",abs)
print(abs)
abs_count=abs.split()
print(len(abs_count))
```

COUNTING THE ARTICLE BODY TEXT

```python
body_content=soup.find_all("div",class_="main-content")
pattern = "<section.*?>.*?</section.*?>"
match_results = re.findall(pattern, html, re.IGNORECASE)
body=' '
for i in match_results:
  body=body+' '+i

body = re.sub("<.*?>", "",body)
body=body.split()
len(body)
```

COUNTING THE IMAGES

```python
image_count=len(soup.find_all("img"))
print(image_count)
```

EXTRACTING REFRENCES

```python
references=html.find("<div id=\"MagazineFulltextArticleBodySuffix\">")
reference_end=html.find("Download references")
end=html[references:reference_end]
pattern = "<p.*?>.*?</p.*?>"

match_results = re.findall(pattern, end, re.IGNORECASE)
reference=' '
for i in match_results:
  reference=reference+'\n'+i

reference = re.sub("<.*?>"," ",reference)
print(reference)
```

PAPER_STATS CSV

```python
dictionary={"Title":title,"Abstract":abs,"Abstract Count":len(abs_count),"Body Count":len(body)}
paper=pd.DataFrame(dictionary,index=[0])
paper.to_csv(r"D:\Intern Files\Programming challenge\paper_stats.csv")

```

REFERENCES CSV

```python
references=html.find("<div id=\"MagazineFulltextArticleBodySuffix\">")
reference_end=html.find("Download references")
end=html[references:reference_end]
pattern = "<p.*?>.*?</p.*?>"

match_results = re.findall(pattern, end, re.IGNORECASE)
reference=[]
for i in match_results:
  reference.append(re.sub("<.*?>"," ",i))

df=pd.DataFrame({},columns=['Reference','Links'])
link=[]
ref=[]
for i in reference:
  ref.append(i)
  link.append(re.findall('https://\S+',i))
df['Links']=link
df['Reference']=ref
```

```python
print(df)
df.to_csv(r"D:\Intern Files\Programming challenge\references.csv")
```
